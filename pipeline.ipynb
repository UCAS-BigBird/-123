```python
import pandas as pd
from sklearn.preprocessing import  StandardScaler,RobustScaler
from sklearn.linear_model import  LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.svm import SVR
from sklearn import metrics

c=pd.read_table(r'C:\Users\UCAS_BigBird\Desktop\zhengqi_train.txt')
d=pd.read_table('C:/Users/UCAS_BigBird/Desktop/zhengqi_test.txt')
train_y=c['target']
train_x=c.drop(['target'],axis=1,inplace=True)

import numpy as np
cc=np.linspace(0.01,10,30)
mse_total=[]
for i in range(len(cc)):
    clf2=make_pipeline(StandardScaler(),
                   SVR(gamma='scale', C=cc[i], epsilon=0.2))
    clf2.fit(c,train_y)
    final_c=clf2.predict(c)
    mse=metrics.mean_squared_error(final_c,train_y)
    mse_total.append(mse)
    print("第{}轮结束.\n".format(i))
# clf=make_pipeline(StandardScaler(),LinearRegression())
# clf1=make_pipeline(LinearRegression())
# clf2=make_pipeline(StandardScaler(),
#                    SVR(gamma='scale', C=1.0, epsilon=0.2))

# clf.fit(c,train_y)
# clf1.fit(c,train_y)
# clf2.fit(c,train_y)
```
